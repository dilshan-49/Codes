```python
import numpy as np
import matplotlib.pyplot as plt
def fn(t,N):
    fnt=0
    for i in range(0,N):
        k=2*i+1
        if k%4==1:
            uk=4*np.cos(2*np.pi*k*t)/(k*np.pi)
        else:
            uk=-4*np.cos(2*np.pi*k*t)/(k*np.pi)
        fnt+=uk
    return fnt

def stepfn(t,N):
    fnt=0
    for i in range(0,N):
        k=2*i+1
        uk=4*np.sin(2*np.pi*k*t)/(k*np.pi)
        fnt+=uk
    return fnt
```

```python
t=np.arange(-1,1,1/200)
shit_stept=fn(t,50)
stept=stepfn(t,50)

fig,ax=plt.subplots(figsize=(5,5))
ax.plot(t,stept)
ax.plot(t,shit_stept)
plt.show
```
